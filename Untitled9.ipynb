```python
import pydicom
import numpy as np
import matplotlib.pyplot as plt

def load_dicom_and_remove_couch(dicom_path, crop_ratio=0.35):
    # Step 1: Load DICOM
    ds = pydicom.dcmread(dicom_path)
    image = ds.pixel_array.astype(np.float32)

    # Step 2: Normalize to [0, 1] for display
    image_norm = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Step 3: Calculate crop height and crop the couch
    height = image_norm.shape[0]
    crop_height = int(height * (1 - crop_ratio))
    image_no_couch = image_norm[:crop_height, :]

    # Step 4: Plot original and couch-removed image
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image_norm, cmap='gray')
    plt.title("Original DICOM (Normalized)")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(image_no_couch, cmap='gray')
    plt.title(f"Couch Removed (Bottom {int(crop_ratio * 100)}%)")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    return image_norm, image_no_couch

# ---- CALL THE FUNCTION ----
dicom_path = "/content/2013030702202100000.0.0.14 - Lung - Althaf.dcm"
original_image, cropped_image = load_dicom_and_remove_couch(dicom_path)
```

```python
import pydicom
import numpy as np
import matplotlib.pyplot as plt

def load_dicom_and_remove_couch_centered(dicom_path, crop_ratio=0.35):
    # Load DICOM
    ds = pydicom.dcmread(dicom_path)
    image = ds.pixel_array.astype(np.float32)

    # Normalize to [0, 1]
    image_norm = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Crop the bottom part (couch)
    original_height, width = image_norm.shape
    crop_height = int(original_height * (1 - crop_ratio))
    image_cropped = image_norm[:crop_height, :]

    # Create black canvas (same size as original)
    centered_image = np.zeros_like(image_norm)

    # Calculate vertical offset to center cropped image
    offset_y = (original_height - crop_height) // 2

    # Place the cropped image in the vertical center
    centered_image[offset_y:offset_y + crop_height, :] = image_cropped

    # Plot original and centered cropped image
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image_norm, cmap='gray')
    plt.title("Original DICOM")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(centered_image, cmap='gray')
    plt.title("Couch Removed (Centered)")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    return image_norm, centered_image
```

```python
import pydicom
import numpy as np
from PIL import Image
from skimage import img_as_ubyte
import os
import matplotlib.pyplot as plt

def convert_dicom_to_png(dicom_path, output_folder, size=(128, 128)):
    # Load and normalize DICOM image
    ds = pydicom.dcmread(dicom_path)
    image = ds.pixel_array.astype(np.float32)
    image_norm = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Convert to uint8 and resize
    image_uint8 = img_as_ubyte(image_norm)
    image_pil = Image.fromarray(image_uint8)
    image_resized = image_pil.resize(size, Image.BILINEAR)

    # Ensure output folder exists
    os.makedirs(output_folder, exist_ok=True)
    output_path = os.path.join(output_folder, "dicom_converted_128x128.png")

    # Save the image
    image_resized.save(output_path)
    print(f"✅ Saved PNG to: {output_path}")

    # Show the image
    plt.imshow(image_resized, cmap='gray')
    plt.title("Converted DICOM to PNG (128×128)")
    plt.axis("off")
    plt.show()

    return output_path
```

```python
dicom_path = "/content/2013030702202100000.0.0.14 - Lung - Althaf.dcm"
output_folder = "/content/converted_pngs"

convert_dicom_to_png(dicom_path, output_folder)
```

```python
import pydicom
import numpy as np
from PIL import Image
from skimage import img_as_ubyte
import os
import matplotlib.pyplot as plt

def remove_couch_and_convert_to_png(dicom_path, output_folder, crop_ratio=0.35, size=(128, 128)):
    # Step 1: Load and normalize DICOM
    ds = pydicom.dcmread(dicom_path)
    image = ds.pixel_array.astype(np.float32)
    image_norm = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Step 2: Remove couch (crop bottom)
    height, width = image_norm.shape
    crop_height = int(height * (1 - crop_ratio))
    image_cropped = image_norm[:crop_height, :]

    # Step 3: Center cropped image back into full frame
    centered_image = np.zeros_like(image_norm)
    offset_y = (height - crop_height) // 2
    centered_image[offset_y:offset_y + crop_height, :] = image_cropped

    # Step 4: Convert to uint8, resize to 128×128
    image_uint8 = img_as_ubyte(centered_image)
    image_pil = Image.fromarray(image_uint8)
    image_resized = image_pil.resize(size, Image.BILINEAR)

    # Step 5: Save to folder
    os.makedirs(output_folder, exist_ok=True)
    output_path = os.path.join(output_folder, "couch_removed_128x128.png")
    image_resized.save(output_path)
    print(f"✅ Saved couch-removed PNG: {output_path}")

    # Step 6: Display the image
    plt.imshow(image_resized, cmap='gray')
    plt.title("Couch-Removed Image (128×128)")
    plt.axis("off")
    plt.show()

    return output_path
```

```python
import cv2
from skimage import exposure
from sklearn.preprocessing import StandardScaler

def apply_normalizations(image_path, output_folder):
    # Load grayscale image
    image = Image.open(image_path).convert("L")
    image_np = np.array(image).astype(np.float32)

    os.makedirs(output_folder, exist_ok=True)

    # ------------------ 1. CLAHE ------------------
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    image_clahe = clahe.apply(image_np.astype(np.uint8))
    Image.fromarray(image_clahe).save(os.path.join(output_folder, "clahe.png"))

    # ------------------ 2. Min-Max ------------------
    min_val = np.min(image_np)
    max_val = np.max(image_np)
    image_minmax = ((image_np - min_val) / (max_val - min_val)) * 255
    image_minmax_uint8 = image_minmax.astype(np.uint8)
    Image.fromarray(image_minmax_uint8).save(os.path.join(output_folder, "minmax.png"))

    # ------------------ 3. Z-Score ------------------
    flat = image_np.flatten().reshape(-1, 1)
    scaler = StandardScaler()
    flat_scaled = scaler.fit_transform(flat).reshape(image_np.shape)

    # Rescale to 0-255 for viewing
    zscore_normalized = ((flat_scaled - flat_scaled.min()) /
                         (flat_scaled.max() - flat_scaled.min())) * 255
    image_zscore = zscore_normalized.astype(np.uint8)
    Image.fromarray(image_zscore).save(os.path.join(output_folder, "zscore.png"))

    # ------------------ Display All ------------------
    titles = ["CLAHE", "Min-Max", "Z-Score"]
    images = [image_clahe, image_minmax_uint8, image_zscore]

    plt.figure(figsize=(12, 4))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, 3, i+1)
        plt.imshow(img, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.suptitle("Normalization Techniques (128×128)", fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
input_image = "/content/converted_pngs/couch_removed_128x128.png"
output_norm_folder = "/content/normalized_outputs"

apply_normalizations(input_image, output_norm_folder)
```

```python
from skimage.filters import unsharp_mask

def apply_unsharp_masking(image_path, output_path, radius=1.0, amount=1.5):
    # Load image (grayscale)
    image = Image.open(image_path).convert('L')
    image_np = np.array(image).astype(np.float32) / 255.0  # normalize to 0–1

    # Apply Unsharp Masking
    sharpened = unsharp_mask(image_np, radius=radius, amount=amount)

    # Rescale to 0–255 and convert to uint8
    sharpened_uint8 = (sharpened * 255).clip(0, 255).astype(np.uint8)
    Image.fromarray(sharpened_uint8).save(output_path)
    print(f" Unsharp masked image saved to: {output_path}")

    # Show original and sharpened image side-by-side
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title("Original (Couch-Removed)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(sharpened_uint8, cmap='gray')
    plt.title("Unsharp Masked")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
input_image = "/content/converted_pngs/couch_removed_128x128.png"
output_denoised_path = "/content/converted_pngs/unsharp_masked_128x128.png"

apply_unsharp_masking(input_image, output_denoised_path)
```

```python
from torchvision.transforms import functional as TF
from PIL import Image
import numpy as np
import os
import random
import matplotlib.pyplot as plt

def better_augmentation(image_path, output_folder, num_augmented=5):
    os.makedirs(output_folder, exist_ok=True)

    base_image = Image.open(image_path).convert('L')
    base_np = np.array(base_image)

    augmented_images = []

    for i in range(num_augmented):
        img = base_image.copy()

        # Horizontal flip
        if random.random() > 0.5:
            img = TF.hflip(img)

        # Slight rotation ±10°
        angle = random.uniform(-10, 10)
        img = TF.rotate(img, angle, fill=0)

        # Slight zoom via center crop and resize
        scale = random.uniform(0.9, 1.0)
        crop_size = int(128 * scale)
        left = (128 - crop_size) // 2
        top = (128 - crop_size) // 2
        img = img.crop((left, top, left + crop_size, top + crop_size))
        img = img.resize((128, 128), Image.BILINEAR)

        # Add low Gaussian noise
        img_np = np.array(img).astype(np.float32)
        noise = np.random.normal(0, 4, img_np.shape)
        img_np += noise
        img_np = np.clip(img_np, 0, 255).astype(np.uint8)
        img = Image.fromarray(img_np)

        # Slight brightness variation
        brightness_factor = random.uniform(0.9, 1.1)
        img = TF.adjust_brightness(img, brightness_factor)

        # Save and collect for preview
        save_path = os.path.join(output_folder, f"aug_better_{i+1}.png")
        img.save(save_path)
        augmented_images.append(img)

    print(f"✅ {num_augmented} cleanly augmented images saved to: {output_folder}")

    # Display all augmented images
    plt.figure(figsize=(15, 3))
    for i, img in enumerate(augmented_images):
        plt.subplot(1, num_augmented, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f"Aug {i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
input_image = "/content/converted_pngs/unsharp_masked_128x128.png"
output_aug_folder = "/content/better_augmented"

better_augmentation(input_image, output_aug_folder, num_augmented=5)
```

```python
def rotation_augmentation(image_path, output_folder, angles=[0, 45, 90, 120, 180]):
    os.makedirs(output_folder, exist_ok=True)

    base_image = Image.open(image_path).convert("L")
    rotated_images = []

    for angle in angles:
        rotated = TF.rotate(base_image, angle, fill=0)
        save_path = os.path.join(output_folder, f"rotated_{angle}deg.png")
        rotated.save(save_path)
        rotated_images.append((rotated, angle))

    print(f"✅ {len(angles)} rotated images saved in: {output_folder}")

    # Show all rotations
    plt.figure(figsize=(15, 3))
    for i, (img, angle) in enumerate(rotated_images):
        plt.subplot(1, len(angles), i+1)
        plt.imshow(img, cmap='gray')
        plt.title(f"{angle}°")
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
input_image = "/content/converted_pngs/unsharp_masked_128x128.png"
output_rot_folder = "/content/rotated_augmented"

rotation_augmentation(input_image, output_rot_folder, angles=[0, 45, 90, 120, 180])
```

```python
import os
import random
from PIL import Image, ImageFilter
import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt

def full_data_augmentation(image_path, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    # Load original image as grayscale (128x128 assumed)
    base_image = Image.open(image_path).convert('L')

    # Define transformations
    rotation_angles = [0, 45, 90, 120, 180]
    blur_options = [None, 2]  # Gaussian blur radius
    flip_options = [None, 'horizontal']

    preview_images = []
    count = 0

    for angle in rotation_angles:
        # Rotate
        rotated = TF.rotate(base_image, angle, fill=0)

        for flip in flip_options:
            if flip == 'horizontal':
                flipped = TF.hflip(rotated)
            else:
                flipped = rotated

            for blur_radius in blur_options:
                final_img = flipped.copy()

                if blur_radius:
                    final_img = final_img.filter(ImageFilter.GaussianBlur(radius=blur_radius))

                # Save augmented image
                filename = f"aug_rot{angle}_{flip or 'noflip'}_blur{blur_radius or 'none'}.png"
                filepath = os.path.join(output_folder, filename)
                final_img.save(filepath)

                preview_images.append((final_img, filename))
                count += 1

    print(f"\n✅ Saved {count} augmented images to: {output_folder}\n")

    # --- Show a few as preview ---
    plt.figure(figsize=(15, 6))
    for i, (img, name) in enumerate(preview_images[:6]):
        plt.subplot(2, 3, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(name, fontsize=8)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
input_img = "/content/converted_pngs/unsharp_masked_128x128.png"
aug_output_folder = "/content/full_augmented_dataset"

full_data_augmentation(input_img, aug_output_folder)
```
